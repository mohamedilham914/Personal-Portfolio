```python
# Drop unnecessary columns
columns_to_drop = ['url', 'timedelta']  # Only drop if they exist
data = data.drop(columns=[col for col in columns_to_drop if col in data.columns])
```

```python
#check for missing values
print(data.isnull().sum())

#fill missing values with the column mean
data - data.fillna(data.mean())
```

```python
# Strip whitespace and standardize column names
data.columns = data.columns.str.strip().str.lower()

# Check the updated column names
print(data.columns)
```

```python
upper_limit = data['shares'].quantile(0.95)
data['shares'] = np.where(data['shares'] > upper_limit, upper_limit, data['shares'])
```

```python
sns.histplot(data=data, x='shares', kde=True)
plt.title('Distribution of Shares', fontsize=16)
plt.xlabel('Shares' , fontsize=14)
plt.ylabel('Frequency' , fontsize=14)
plt.show()
```

```python
sns.histplot(data['shares'], kde = True)
plt.title('Distribution of Shares after Outlier Handling')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['shares']] = scaler.fit_transform(data[['shares']])

print(data[['shares']].head())
```

Encode Categorical Variables

```python
categorical_columns = ['data_channel_is_lifestyle', 'data_channel_is_entertainment',
                       'data_channel_is_bus', 'data_channel_is_socmed',
                       'data_channel_is_tech', 'data_channel_is_world']

data = pd.get_dummies(data, columns=categorical_columns)

# Verify encoded columns
print(data.head())
```

```python
# Strip whitespace and convert column names to lowercase
data.columns = data.columns.str.strip().str.lower()

# Check the updated column names
print(data.columns)
```

```python
# Check which data channel columns are present
data_channels = [col for col in ['data_channel_is_lifestyle', 'data_channel_is_entertainment',
                                 'data_channel_is_bus', 'data_channel_is_socmed',
                                 'data_channel_is_tech', 'data_channel_is_world'] if col in data.columns]

print("Available data channel columns:", data_channels)
```

```python
# Identify columns for the "1.0" categories
data_channels = [col for col in data.columns if col.endswith('_1.0')]

# Verify the selected columns
print("Selected data channels:", data_channels)
```

```python
# Calculate total shares for each channel
shares_by_channel = {channel: (data[channel] * data['shares']).sum() for channel in data_channels}

# Convert to a Pandas Series for easy plotting
shares_by_channel = pd.Series(shares_by_channel)
print(shares_by_channel)
```

```python
shares_by_channel.plot(kind='bar', color='skyblue')
plt.title('Shares by Data Channel')
plt.xlabel('Data Channel')
plt.ylabel('Total Shares')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot distribution of global sentiment polarity
sns.histplot(data['global_sentiment_polarity'], kde=True)
plt.title('Distribution of Sentiment Polarity')
plt.show()
```
